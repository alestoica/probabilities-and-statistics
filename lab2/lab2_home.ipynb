```python
from random import randint
```

```python
def simulationGroup23():
    birthdays = []
    for pers in range(23):
        birthday = randint(1, 365)
        birthdays.append(birthday)
    return len(birthdays) != len(set(birthdays))
```

```python
nrSimulations = 100000

nrSuccesses = 0;
for s in range(nrSimulations):
    nrSuccesses += simulationGroup23()

probability = nrSuccesses / nrSimulations
```

```python
print("Probability of at least to people in a group of 23 having the same birthday is approximately: " + str(probability))
```

```python
def sameBirthdayProbability():
    uniqueProbability = 1.0
    for i in range(23):
        uniqueProbability *= (365 - i) / 365
    atLeastTwoProbability = 1 - uniqueProbability
    return atLeastTwoProbability

probability = sameBirthdayProbability()
```

```python
print("Probability of at least to people in a group of 23 having the same birthday is approximately: " + str(probability))
```

```python
def generatedPoints(N):
    pointsInCircle = 0
    for point in range(N):
        x = 2 * random() - 1 # generates coordinates x in interval [-1, 1]
        y = 2 * random() - 1 # generates coordinates y in interval [-1, 1]
        if dist((x, y), (0, 0)) <= 1: 
            pointsInCircle += 1
        plot(x, y, 'ro')
    axis('equal')
    return pointsInCircle / N


N = 500  # can also be 1000 or 2000
relativeFrequency = generatedPoints(N)
```

```python
from random import random
from math import dist
```

```python
def generatedPoints(N):
    pointsCenter = 0

    for _ in range(N):
        x = random()  # generates coordinates x in interval [0, 1]
        y = random()  # generates coordinates x in interval [0, 1]

        distanceCenter = dist((x, y), (0.5, 0.5))
        distanceVertices = min([dist((x, y), (0, 0)), dist((x, y), (1, 0)), dist((x, y), (0, 1)), dist((x, y), (1, 1))])

        if distanceCenter < distanceVertices:
            pointsCenter += 1

    relativeFrequency = pointsCenter / N

    return relativeFrequency


N = 500  # can also be 1000 and 2000
relativeFrequency = generatedPoints(N)
```

```python
print("The relative frequency of points closer to the center of the circle: " + str(relativeFrequency))
```
