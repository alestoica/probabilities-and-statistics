```python
from random import randint

def simulations(nrPers):
    birthdays = []
    for i in range(nrPers):
        birthday = randint(1, 365)
        birthdays.append(birthday)
    return len(birthdays) != len(set(birthdays))

N = 100000

favourableCase = 0
for i in range(N):
    favourableCase += simulations(26)

prob = favourableCase / N
print("Probability = " + str(prob))
```

```python
from math import pow

def probability(nrPers):
    defavCase = 1
    
    for i in range(nrPers):
        defavCase *= (365 - i + 1)

    favCase = pow(365, nrPers)

    return 1 - defavCase / favCase

print("Probability = " + str(probability(26)))
```

```python
from random import random
from math import dist

def pointsInCircle(nr_points):
    nr_points_in_circle = 0
    O = [0.5, 0.5]
    axis('square')
    axis((0, 1, 0, 1))
    for p in range(nr_points):
        x = random()
        y = random()
        P = [x, y]
        distPO = dist(P, O)
        if distPO < 0.5:
            nr_points_in_circle += 1
            plot(x, y, '*m')
    return nr_points_in_circle

N = 1000

nr = pointsInCircle(N)

prob = nr / N

print("Probability = " + str(prob))
```
