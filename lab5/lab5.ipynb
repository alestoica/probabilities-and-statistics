```python
from scipy.stats import uniform
from numpy import cumsum

N = 1000
x = [1, 2, 3, 4]
p = [0.46, 0.4, 0.1, 0.04]

q = cumsum(p)
result = []

for i in range(1, N):
    u = uniform.rvs()
    j = 0
    while u > q[j]:
        j += 1
    result.append(x[j])

print(result)
```

```python
from matplotlib.pyplot import bar, hist, grid, show, legend

data = [i for i in result]

bin_edges = [k + 0.5 for k in range(5)]

hist(data, bin_edges, density = True, rwidth = 0.9, color = 'magenta', edgecolor = 'black', alpha = 0.5, label = 'frecvente relative')

show()
```

```python
from scipy.stats import uniform
from math import log

N = 1000

alpha = 1/12
result = []

for i in range(N):
    u = uniform.rvs()
    x = -1/alpha * log(1 - u)
    result.append(x)

print(result)
```

```python
from matplotlib.pyplot import show, hist, grid, legend, xticks, plot
from scipy.stats import expon
from math import log

alpha = 1/12
data = result
hist(data, bins=12, range=(0, 60), density=True)

x = range(60)
plot(x, expon.pdf(x, loc=0, scale=1/alpha), 'magenta')
xticks(range(0, 60, 5))
grid()
show()
```
