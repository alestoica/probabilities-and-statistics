```python
import random
from math import comb, perm

def simulations(nrSimulations):
    countA = 0
    countAB = 0
    
    list = ['r'] * 5 + ['a'] * 3 + ['v'] * 2

    for i in range(nrSimulations):
        chosen = random.sample(list, 3)
        if 'r' in chosen:
            countA += 1
            if chosen == ['r', 'r', 'r']:
                countAB += 1

    return countAB/countA

print(simulations(10000))

```

```python
from random import randrange
from matplotlib.pyplot import bar, hist, grid, show, legend

data = [randrange(1,7) for _ in range(500)]

bin_edges = [k + 0.5 for k in range(7)]

hist(data, bin_edges, density = True, rwidth = 0.9, color = 'green', edgecolor = 'black', alpha = 0.5, label = 'frecvente relative')

distribution = dict([(i,1/6) for i in range(1,7)])

bar(list(distribution.keys()), 1/6, width = 0.85, color = 'red', edgecolor = 'black', alpha= 0.6, label = 'probabilitati')

legend(loc = 'lower left')
grid()
show()
```
