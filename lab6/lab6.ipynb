```python
from scipy.stats import norm
from numpy import mean, std, linspace
from matplotlib.pyplot import show, hist, grid, legend, xticks, plot

N = 1000
data = []
m = 165
sigma = 10 

data = norm.rvs(loc=m, scale=sigma, size = N)

hist(data, 16, density = True, rwidth = 0.9, color = 'magenta', edgecolor = 'black', range = (130, 210))
x = linspace(130, 210, 10000)
plot(x, norm.pdf(x, m, sigma))
grid()

print("media: ", mean(data))
print("deviatia standard: ", std(data))
print("proportia de valori in intervalul [160, 170]: ", len([x for x in data if 160 <= x <= 170])/N)

norm.cdf(170, loc = m, scale = sigma) - norm.cdf(160, loc = m, scale = sigma)
norm((160 < data)&(data < 170))

show()
```
